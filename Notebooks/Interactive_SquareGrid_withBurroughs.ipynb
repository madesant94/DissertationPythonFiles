```python
import networkx as nx
import osmnx as ox
from shapely.geometry import LineString, mapping
import geopandas as gpd
import pandas as pd
from ipyleaflet import *

import osmnx as ox

import math

def get_bbox(point, dist=4500):
    
    earth_radius = 6371000 #meters
    lat, lng = point
    
    delta_lat = (dist / earth_radius) * (180 / math.pi)
    delta_lng = (dist / earth_radius) * (180 / math.pi) / math.cos(lat * math.pi / 180)
    
    north  = lat + delta_lat
    south  = lat - delta_lat
    east = lng + delta_lng
    west = lng - delta_lng
    return north, south, east, west

location_point = (51.490938, -0.196054) # Center point used throughout the project for Graph size 6.5 km

north, south, east, west = get_bbox(location_point, 3250)

point1 = (north, east)
point2 = (north, west)
point3 = (south, west)
point4 = (south, east)
mainPoly = [point1,point2,point3,point4,point1]

print("polyLimits", mainPoly)

# Get the burroughs data
place_name = 'City of Westminster, Greater London, UK'
polygon = ox.geocode_to_gdf(place_name)
polygon_ = polygon.geometry
coords_list = list(polygon_[0].exterior.coords)
coords_list_Wesminster = []
for i in coords_list:
    coords_list_Wesminster.append((i[1],i[0]))

# Get the burroughs data
place_name = 'Kensington and Chelsea, Greater London, UK'
polygon = ox.geocode_to_gdf(place_name)
polygon_ = polygon.geometry
coords_list = list(polygon_[0].exterior.coords)
coords_list_Kensington = []
for i in coords_list:
    coords_list_Kensington.append((i[1],i[0]))

# Get the burroughs data    
place_name = 'Hammersmith and Fulham, Greater London, UK'
polygon = ox.geocode_to_gdf(place_name)
polygon_ = polygon.geometry
coords_list = list(polygon_[0].exterior.coords)
coords_list_Hammersmith = []
for i in coords_list:
    coords_list_Hammersmith.append((i[1],i[0]))

# create map
m = Map(center=location_point, zoom=13, basemap=basemaps.CartoDB.Positron)

polygonMain = Polygon(
        locations= mainPoly,
        color="steelblue",
        fill_opacity= 0.1,
        fill_color="steelblue",
        weight = 3
    )

# Add square of interest
m.add_layer(polygonMain)

# create polygon for burrough
polygonWestminster = Polygon(
        locations= coords_list_Wesminster,
        color="salmon",
        fill_opacity= 0.0,
        fill_color="salmon",
        weight = 2
    )

polygonKensington = Polygon(
        locations= coords_list_Kensington,
        color="salmon",
        fill_opacity= 0.0,
        fill_color="salmon",
        weight = 2
    )

polygonHammersmith = Polygon(
        locations= coords_list_Hammersmith,
        color="salmon",
        fill_opacity= 0.0,
        fill_color="salmon",
        weight = 2
    )


marker = Marker(location=location_point, draggable=True)
m.add_layer(marker);

m.add_layer(polygonWestminster)
m.add_layer(polygonKensington)
m.add_layer(polygonHammersmith)

m.layout.width = '800px'  # Adjust the desired width
m.layout.height = '800px'
m.save('polygonCity.html')

print('done')
```
