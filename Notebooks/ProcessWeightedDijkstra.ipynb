## Plot main boxplots (shortest / cleanest)

```python
fig, ax = plt.subplots(figsize=(6,4))
from matplotlib.patches import Patch
red_circle = dict(markerfacecolor='white', marker='o', markeredgecolor='black', markersize = 5)
medianprops = dict(linestyle='-', linewidth=2, color='black')

# group data
data = [shortestSmallPollution, cleanestSmallPollution,
                 shortestMediumPollution, cleanestMediumPollution,
                 shortestLargePollution, cleanestLargePollution,
                
                ]

labels = ['Cleanest', 'Shortest', 'Cleanest', 'Shortest', 'Cleanest', 'Shortest']


bp = ax.boxplot(data,
             medianprops = medianprops, showfliers=False,labels = labels, patch_artist=True)

  
fill_color = ['darkgray','darkgray', 'darkgray','darkgray']

# style elements of boxplot 
for element in ['whiskers', 'fliers', 'caps']:

    n = 0 
    for i in bp[element]:
        plt.setp(i, color="black", linewidth=1.5)
        n = n+1
        
        
        
fill_color = ['lightgrey', 'forestgreen','lightgrey','forestgreen', 'lightgrey','forestgreen']
n = 0
for patch in bp['boxes']:
        print(patch)
        patch.set(facecolor=fill_color[n], edgecolor = 'black', linewidth = 1.5,  alpha = 0.75) 
        n = n+1
        
        
legend_handles = [Patch(facecolor='lightgrey', edgecolor='black', label='Shortest'),
                  Patch(facecolor='forestgreen', edgecolor='black', label='Cleanest')]

# Add the legend
legend = ax.legend(handles=legend_handles, handlelength=2)

plt.grid(True)
#ax1.set_title('Weighted beta = 0.5 Route - % Difference in Pollution')
ax.set_xticklabels(["Shortest", "Cleanest","Shortest" , "Cleanest", "Shortest" , "Cleanest"])
ax.set_ylabel('Sum of Pollutants PM$_{2.5}$ ($\mu$g / $m^3$) per route')
ax.set_xlabel('Routes')

tick_positions = np.arange(0, len(data), 1)
tick_positions = [1.5,3.5,5.5]
print(tick_positions)
tick_labels = ['Small Routes', 'Medium Routes', 'Large Routes']

plt.xticks(tick_positions, tick_labels)

plt.title ('Total Sum of Pollutant PM$_{2.5}$ by Route size')
fig.savefig('./Results/AllSizesRoutesPollution.png')
# -----------------------------------

fig, ax = plt.subplots(figsize=(6,4))
from matplotlib.patches import Patch
red_circle = dict(markerfacecolor='white', marker='o', markeredgecolor='black', markersize = 5)
medianprops = dict(linestyle='-', linewidth=2, color='black')

data = [shortestSmallLength, cleanestSmallLength,
                 shortestMediumLength, cleanestMediumLength,
                 shortestLargeLength, cleanestLargeLength,
                
                ]

labels = ['Cleanest', 'Shortest', 'Cleanest', 'Shortest', 'Cleanest', 'Shortest']


bp = ax.boxplot(data,
             medianprops = medianprops, showfliers=False,labels = labels, patch_artist=True)

  
fill_color = ['darkgray','darkgray', 'darkgray','darkgray']

for element in ['whiskers', 'fliers', 'caps']:

    n = 0 
    for i in bp[element]:
        plt.setp(i, color="black", linewidth=1.5)
        n = n+1
        
        
        
fill_color = ['lightgrey', 'forestgreen','lightgrey','forestgreen', 'lightgrey','forestgreen']
n = 0
for patch in bp['boxes']:
        print(patch)
        patch.set(facecolor=fill_color[n], edgecolor = 'black', linewidth = 1.5,  alpha = 0.75) 
        n = n+1
        
        
legend_handles = [Patch(facecolor='lightgrey', edgecolor='black', label='Shortest'),
                  Patch(facecolor='forestgreen', edgecolor='black', label='Cleanest')]

# Add the legend
legend = ax.legend(handles=legend_handles, handlelength=2)

#legend = plt.legend(["Cleanest", "Shortest"], handlelength=3)

plt.grid(True)
#ax1.set_title('Weighted beta = 0.5 Route - % Difference in Pollution')
ax.set_xticklabels(["Shortest", "Cleanest","Shortest" , "Cleanest", "Shortest" , "Cleanest"])
ax.set_ylabel('Total distance [m]')
ax.set_xlabel('Routes')

tick_positions = np.arange(0, len(data), 1)
tick_positions = [1.5,3.5,5.5]
print(tick_positions)
tick_labels = ['Small Routes', 'Medium Routes', 'Large Routes']

plt.xticks(tick_positions, tick_labels)

plt.title ("Total Route Length by Route size")
#fig.savefig('./Results/AllSizesRoutesLength.png')

```
