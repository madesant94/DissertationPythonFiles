```python
import pandas as pd
import requests

df = pd.read_csv('./data/Predictions/55/g60_2023_07_20_08_00_00.csv')

dictionary = {}
for index,row in df.iterrows():
    dictionary[row['code']] = round(row["predicted_pm2.5"],2)
            
print(dictionary)

new_dictionary = {'value': dictionary}

url = "https://cleanestpath-gcr.nw.r.appspot.com/data"
#url = "https://cleanestpath.nw.r.appspot.com/data"
response = requests.post(url, json=new_dictionary)
print(response.json())
```
